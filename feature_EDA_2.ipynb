## Initial Feature Selection after EDA

```python
new_df_names = list(new_df.columns)
variables_to_keep = [
    "ListingCreationDate",  # The date the listing was created
    "BorrowerAPR",          # The Borrower's Annual Percentage Rate (APR) for the loan
    "BorrowerRate",         # The Borrower's interest rate for this loan
    "LenderYield",          # The Lender yield on the loan
    "EstimatedEffectiveYield", # Effective yield on the loan after fees and estimated losses
    "EstimatedLoss",          # Estimated principal loss on charge-offs
    "EstimatedReturn",        # Estimated return on the loan
    "ProsperScore",           # A custom risk score
    "ListingCategory",        # The category of the listing that the borrower selected
    "EmploymentStatus",       # The employment status of the borrower
    "EmploymentStatusDuration", # The length in months of the employment status
    "IsBorrowerHomeowner",    # If the borrower is classified as a homeowner
    "CreditScoreRangeLower",  # The lower bound of the borrower's credit score
    "CreditScoreRangeUpper",  # The upper bound of the borrower's credit score
    "CurrentCreditLines",     # Number of current credit lines
    "OpenCreditLines",        # Number of open credit lines
    "TotalCreditLinespast7years", # Number of credit lines in the past seven years
    "OpenRevolvingAccounts",  # Number of open revolving accounts
    "OpenRevolvingMonthlyPayment", # Monthly payment on revolving accounts
    "InquiriesLast6Months",   # Number of inquiries in the past six months
    "AmountDelinquent",       # Dollars delinquent
    "DelinquenciesLast7Years", # Number of delinquencies in the past 7 years
    "PublicRecordsLast10Years", # Number of public records in the past 10 years
    "PublicRecordsLast12Months", # Number of public records in the past 12 months
    "RevolvingCreditBalance", # Dollars of revolving credit
    "BankcardUtilization",    # Percentage of available revolving credit utilized
    "AvailableBankcardCredit", # Total available credit via bank card
    "TradesNeverDelinquent",  # Number of trades never delinquent
    "TradesOpenedLast6Months", # Number of trades opened in the last 6 months
    "DebtToIncomeRatio",      # Debt to income ratio of the borrower
    "IncomeRange",            # The income range of the borrower
    "IncomeVerifiable",       # If the borrower has documentation to support their income
    "StatedMonthlyIncome",    # The monthly income the borrower stated
    "TotalProsperLoans",      # Number of Prosper loans the borrower has
    "OnTimeProsperPayments",  # Number of on time payments on Prosper loans
    "ProsperPaymentsLessThanOneMonthLate", # Payments less than one month late on Prosper loans
    "ProsperPaymentsOneMonthPlusLate", # Payments more than one month late on Prosper loans
    "ProsperPrincipalOutstanding", # Principal outstanding on Prosper loans
    "ScorexChangeAtTimeOfListing", # Credit score change at the time of listing
    "LoanCurrentDaysDelinquent", # The number of days the loan is delinquent
    "LoanOriginalAmount",        # The origination amount of the loan
    "LP_CustomerPayments",       # Gross payments made by the borrower on the loan
    "LP_CustomerPrincipalPayments", # Principal payments made by the borrower on the loan
    "PercentFunded",             # Percent the listing was funded
    "Recommendations",           # Number of recommendations for the borrower
    "InvestmentFromFriendsCount", # Number of friends that invested in the loan
    "InvestmentFromFriendsAmount", # Amount invested by friends
    "Investors",# The number of investors that funded the loan
    "LoanComplete"
]
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
#Apply PCA
pca = PCA(n_components=0.95) # n_components as 0.95 means keep 95% of variance
principal_components = pca.fit_transform(scaled_data)

# The fitted PCA object now contains the principal components
# that together explain up to 95% of the variance.
# principal_components is a new data matrix of reduced dimensions
```

```python
# If you're interested in how many components were selected:
n_components_selected = pca.n_components_

print(f"Number of components selected: {n_components_selected}")
```

```python
#Scree plot

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Number of Components')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1))
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(14, 7))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, color='blue', alpha=0.7)
plt.title('PCA Variance Explained by Component')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Will retain only the first 4 components of the data as it contributes the most significantly to the variance.

```python
#Change number of components to 4
pca = PCA(n_components=4)
principal_components = pca.fit_transform(scaled_data)
```
